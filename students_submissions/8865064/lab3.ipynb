```python
# Import necessary libraries
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import datasets
import matplotlib.pyplot as plt  
```

```python
# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

#body mass index column
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Splitting the data into sets for training/testing 
X_train, X_test, y_train, y_test = train_test_split(diabetes_X, diabetes_y, test_size=0.2, random_state=27)

```

```python
# Create linear regression object
model = LinearRegression()
# Train the model using the training sets
model.fit(X_train, y_train)
```

```python
# Make predictions using the training set
y_pred_Train = model.predict(X_train)
# Make predictions using the testing set
y_pred_Test = model.predict(X_test)
```

```python
# The coefficients
print("Coefficients: ", model.coef_)
# The intercept
print("Intercept: ", model.intercept_)

# The mean absolute error of Train data set
print("Mean aboslute error of Training data set: %.2f" % metrics.mean_absolute_error(y_train, y_pred_Train))
# The mean absolute error of Test data set
print("Mean aboslute error of Test data set: %.2f" % metrics.mean_absolute_error(y_test, y_pred_Test))

# The mean squared error of Train data set
print("Mean squared error of Training data set: %.2f" % metrics.mean_squared_error(y_train, y_pred_Train))
# The mean squared error of Test data set
print("Mean squared error of Testing data set: %.2f" % metrics.mean_squared_error(y_test, y_pred_Test))
```

```python
#Training Data Plots
plt.scatter(X_train, y_train, color="black", label="Training Data")
plt.plot(X_train, y_pred_Train, color="blue", linewidth=3, label="Model")

#Test Data Plots
plt.scatter(X_test, y_test, color="red", label="Testing Data")
plt.plot(X_test, y_pred_Test, color="green", linewidth=3, label="Model")

plt.legend()
plt.title("Linear Regression Model")
plt.xlabel("Body Mass Index (BMI)")
plt.ylabel("Target Value")
plt.show()
```

### This is a good fit as the training and test data are scaterred all across the plot. There are relatively low mean absolute errors and high R-squared values
